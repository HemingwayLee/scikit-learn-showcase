# Understand iris dataset

```python
%matplotlib inline

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import datasets
from sklearn.decomposition import PCA

iris = datasets.load_iris()
 
print(f"shape: {iris.data.shape}, 150 samples, 4 features (petal width/height sepal width/height) for each sample")
print(iris.data[:5])
print(iris.target[:5])
```

# Show dataset using first 2 features

```python
# Take the first 2 features
X = iris.data[:, :2]
y = iris.target

x_min, x_max = X[:, 0].min() - .5, X[:, 0].max() + .5
y_min, y_max = X[:, 1].min() - .5, X[:, 1].max() + .5

plt.figure(2, figsize=(8, 6))
plt.clf()

plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Set1, edgecolor='k')
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')

plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.xticks(())
plt.yticks(())
```

# Show dataset using first 3 features

```python
fig = plt.figure(1, figsize=(8, 6))
ax = Axes3D(fig, elev=-150, azim=110)

ax.scatter(iris.data[:, 0], iris.data[:, 1], iris.data[:, 2], c=y, cmap=plt.cm.Set1, edgecolor='k', s=40)

ax.set_title("3D features of iris dataset")
ax.set_xlabel("1st feature")
ax.w_xaxis.set_ticklabels([])
ax.set_ylabel("2nd feature")
ax.w_yaxis.set_ticklabels([])
ax.set_zlabel("3rd feature")
ax.w_zaxis.set_ticklabels([])

plt.show()
```

# Use dataset using PCA with 3 eigenvectors

```python
fig2 = plt.figure(1, figsize=(8, 6))
ax2 = Axes3D(fig2, elev=-150, azim=110)
X_pca = PCA(n_components=3).fit_transform(iris.data)
ax2.scatter(X_pca[:, 0], X_pca[:, 1], X_pca[:, 2], c=y, cmap=plt.cm.Set1, edgecolor='k', s=40)
ax2.set_title("First three PCA directions")
ax2.set_xlabel("1st eigenvector")
ax2.w_xaxis.set_ticklabels([])
ax2.set_ylabel("2nd eigenvector")
ax2.w_yaxis.set_ticklabels([])
ax2.set_zlabel("3rd eigenvector")
ax2.w_zaxis.set_ticklabels([])

plt.show()
```
